Feature scaling the data 


encoding 

```python
data['godquality']=[1 if x>=7 else 0 for x in data['quality']]
```

```python
X=data.drop(['quality', 'godquality'],axis=1)
y=data['godquality']
```

```python
X.head()
```

```python
y.head()
```

First splitting the data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf_model=RandomForestClassifier(random_state=42)
```

TRAINING DATA WITHIN AND WITHOUT SCALING      


```python
rf_model.fit(X_train,y_train)
```

Prediction 

```python
y_pred=rf_model.predict(X_test)
```

Calculationg Score

```python
rf_accuracy=accuracy_score(y_test,y_pred)
rf_accuracy
```

```python
from sklearn.linear_model import LogisticRegression 
lr_model=LogisticRegression()
lr_model.fit(X_train,y_train) 
y_pred_lr=lr_model.predict(X_test)
lr_accuracy=accuracy_score(y_test,y_pred_lr)
lr_accuracy
```

```python
import joblib

# Lagre Random Forest-modellen
joblib.dump(rf_model, 'wine_quality_model.pkl')
```
